```python
import pandas as pd

dataset = pd.read_csv("HospInfo.csv")
```

```python
dataset.head()
```

```python
lista_hospital = ['Provider ID', 'Hospital Name', 'Address', 'Phone Number',
                  'Hospital Type', 'Hospital Ownership', 'Emergency Services',
                  'Meets criteria for meaningful use of EHRs']
hospital_df = dataset[lista_hospital]
hospital_df.head()
```

```python
lista_estat = ['Provider ID', 'Mortality national comparison', 'Safety of care national comparison',
               'Readmission national comparison', 'Patient experience national comparison',
               'Effectiveness of care national comparison', 'Timeliness of care national comparison',
               'Efficient use of medical imaging national comparison'
                ] 
estatisticas_df = dataset[lista_estat]
estatisticas_df.head()
```

```python
lista_loc = ['Provider ID', 'City', 'State', 'ZIP Code', 'County Name'] 
localizacao_df = dataset[lista_loc]
localizacao_df.head()
```

```python
db.hospital.insert_many(hospital_df.to_dict('records'))
```

```python
db.estatistica.insert_many(estatisticas_df.to_dict('records'))
```

```python
db.localizacao.insert_many(localizacao_df.to_dict('records'))
```

```python
county_name = 'MARSHALL'

provider_ids = db.localizacao.find({"County Name": county_name}, {"Provider ID": 1})

provider_ids = [doc['Provider ID'] for doc in provider_ids]

print(provider_ids)

hospitals_in_county = db.hospital.find({"Provider ID": {"$in": provider_ids}})

hospitals_list = list(hospitals_in_county)

for hospital in hospitals_list:
    print(hospital)
```
